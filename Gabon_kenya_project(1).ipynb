**tranformation of non-tidy columns in the list of dataframe columns**

```python
# Transformation en format tidy
tidy_data = Gabon.melt(
    id_vars=['asdf_id', 'Level', 'gqid', 'id', 'shapeGroup', 'shapeID', 'shapeISO', 'shapeName', 'shapeType'],
    var_name='variable_year',
    value_name='value'
)

# Séparer la colonne 'variable_year' en deux nouvelles colonnes 'variable' et 'year'
tidy_data[['variable', 'year', 'stat']] = tidy_data['variable_year'].str.split('.', expand=True)

# Supprimer les colonnes inutiles après la séparation
tidy_data = tidy_data.drop(columns=['variable_year', 'stat'])

#convertion des annees en numerique

tidy_data['year']=pd.to_numeric(tidy_data['year'])

# Vérifier le résultat final
tidy_data

```
